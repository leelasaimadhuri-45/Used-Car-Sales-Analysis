```python
#%%
url='https://www.spinny.com/used-cars-in-mumbai/s/?filterObject={%22make%22:[%22audi%22,%22bmw%22,%22chevrolet%22,%22citroen%22,%22datsun%22,%22fiat%22,%22force-motors%22,%22ford%22,%22honda%22,%22hyundai%22,%22jaguar%22,%22jeep%22,%22kia%22,%22land-rover%22,%22lexus%22,%22mahindra%22,%22maruti-suzuki%22,%22mercedes-benz%22,%22mg-motors%22,%22nissan%22,%22renault%22,%22skoda%22,%22tata%22,%22toyota%22,%22volkswagen%22]}&referrer=/home/&utm_medium=gads_c_search&utm_source=SPD-Search-Top8-National-Brand-EM-HYD&utm_campaign=Brand-EM&utm_content=690971557108&utm_term=e_spinny&partner_id=CjwKCAiAlrXJBhBAEiwA-5pgwhBKwmUtYUUWGrpBTFmy1iR_q1bpWr7j6KSaAkUV7C1D5pjqIGYCzRoCRw4QAvD_BwE&source=googleads&sub_source=SPD-Search-Top8-National-Brand-EM-HYD&gclid=CjwKCAiAlrXJBhBAEiwA-5pgwhBKwmUtYUUWGrpBTFmy1iR_q1bpWr7j6KSaAkUV7C1D5pjqIGYCzRoCRw4QAvD_BwE'
#%%
from bs4 import BeautifulSoup
import time
#%%
!pip install selenium
from selenium import webdriver
#%%
browser = webdriver.Chrome()
browser.get(url)
time.sleep(3)
#%%
prev_height = browser.execute_script("return document.body.scrollHeight")
#%%
while True:
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3)

    new_height = browser.execute_script("return document.body.scrollHeight")
    if new_height == prev_height:
        break
    prev_height = new_height

html = browser.page_source
browser.quit()
#%%
page_html = BeautifulSoup(html, 'html.parser')
cars = page_html.find_all('div', {'class': "CarListingCardDetail__productDetailContainer"})
print("Total cars found:", len(cars))
#%%
manufacturing_year_list = []
brand_list = []
name_list = []
kms_travelled_list = []
fuel_type_list = []
price_list = []
#%%
for car in cars:
    span = car.find('span')
    name = span.get_text(strip=True) if span else ""
    parts = name.split()

    manufacturing_year_list.append(parts[0] if len(parts) > 0 else None)
    brand_list.append(parts[1] if len(parts) > 1 else None)
    name_list.append(" ".join(parts[1:]) if len(parts) > 1 else None)

    ul = car.find("ul", class_="CarListingCardDetail__more")
    li_list = ul.find_all("li") if ul else []

    kms_travelled_list.append(li_list[0].get_text(strip=True) if len(li_list) > 0 else None)
    fuel_type_list.append(li_list[1].get_text(strip=True) if len(li_list) > 1 else None)

    price = car.find("span", class_="ListingBrandModelDetail__priceWithRupeeSymbol")
    price_list.append(price.get_text(strip=True) if price else None)
#%%
import pandas as pd

df = pd.DataFrame({
    "Model Year": manufacturing_year_list,
    "Brand Name": brand_list,
    "Name": name_list,
    "Distance Driven": kms_travelled_list,
    "Fuel Type": fuel_type_list,
    "Price": price_list
})
```

```python
df["Price"] = df["Price"].str.split(" ").str[0].astype(float)
```

```python
df["Distance Driven"] = df["Distance Driven"].str.extract(r"^(\d+\.?\d*)").astype(float)
df["Distance Driven"] = df["Distance Driven"]*1000
```
